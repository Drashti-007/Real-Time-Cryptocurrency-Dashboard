```python
import requests
import pandas as pd
from datetime import datetime
```

```python
url = "https://api.coingecko.com/api/v3/simple/price?ids=bitcoin,ethereum,solana,dogecoin&vs_currencies=usd"
response = requests.get(url)
```

```python
response.status_code
```

```python
response.json()
```

```python
data = response.json()
```

```python
print("Bitcoin:", data["bitcoin"]["usd"])
print("Ethereum:", data["ethereum"]["usd"])
print("Solana:", data["solana"]["usd"])
print("Dogecoin:", data["dogecoin"]["usd"])
```

```python
current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
print(current_time)
```

```python
crypto_data = {
    "Cryptocurrency": ["Bitcoin", "Ethereum", "Solana", "Dogecoin"],
    "Price (USD)": [
        data["bitcoin"]["usd"],
        data["ethereum"]["usd"],
        data["solana"]["usd"],
        data["dogecoin"]["usd"]
    ],
    "Timestamp": [current_time] * 4
}

df = pd.DataFrame(crypto_data)
```
