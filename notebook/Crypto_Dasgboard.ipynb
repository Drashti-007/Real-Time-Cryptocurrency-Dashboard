```python
import requests
import pandas as pd
from datetime import datetime
```

```python
url = "https://api.coingecko.com/api/v3/simple/price?ids=bitcoin,ethereum,solana,dogecoin&vs_currencies=usd"
response = requests.get(url)
```

```python
response.status_code
```

```python
response.json()
```

```python
data = response.json()
```

```python
crypto_names = []
crypto_prices = []

for crypto in data:
    crypto_names.append(crypto)
    crypto_prices.append(data[crypto]["usd"])

print(crypto_names)
print(crypto_prices)
```

```python
current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

df = pd.DataFrame({
    "Cryptocurrency": crypto_names,
    "Prices (USD)": crypto_prices,
    "Timestamp": [current_time] * len(crypto_names)
})

df
```
